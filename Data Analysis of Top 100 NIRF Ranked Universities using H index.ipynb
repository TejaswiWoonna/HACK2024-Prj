```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:/Users/91897/Desktop/PRJ 1/Vellore Institute of Technology.csv")
df
```

```python
FIN=df.drop(['Authors','Author(s) ID','Volume','Issue','Art. No.','Page start','Page end','Page count','DOI','Link','Document Type','Publication Stage','Open Access','Source','EID'],axis=1)
```

```python
FIN.rename(columns = {'Cited by':'citations'}, inplace = True)
FIN['citations'].replace(np.nan, 0,inplace=True)
```

```python
class Solution:
    def hIndex(self, citations: list[int]) -> int:
        citations.sort(reverse=True)
        
        for indx, citation in enumerate(citations):
            if indx >= citation:
                return indx
        return len(citations)
```

```python
citations = FIN.citations.values.tolist()
print(citations)
```

```python
# Create an instance of the Solution class
solution = Solution()

# Provide a list of citations to the hIndex method
h_index = solution.hIndex(citations)

# Print the calculated H-index
print("H-index:",h_index)
```
